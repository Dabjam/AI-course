```python
import numpy as np

# Cоздаем одномерный массив
a = np.array([1, 2, 3, 4])

# Вывод массива
print(a)

# Тип данных в массиве
print(a.dtype)

# Размерность массива
print(a.shape)
```

```python
import numpy as np

# Cоздаем двумернный массив
b = np.array([[1, 22, 3], [5, 54, 2]])

# Выведем его, узанем тип переменной и узнаем размерность массива
print(b)
print(b.dtype)
print(b.shape)
```

```python
import numpy as np 

# Создаем матрицу состоящую из нулей и размерностью 3x5  
matrix0 = np.zeros((3, 5))

# Создаем две матрицы состоящую из единиц и размерностью 3x4 
matrix1 = np.ones((2, 3, 4))

# Вывод матриц
print(matrix0, matrix1, sep="\n")

```

```python
import numpy as np

# Создаем массив благодаря функции arange от 3 до 9
ls = np.arange(3, 10)

# Вывод массива
print(ls)

# Получение отдельных элементво: вывод первого и последнего элемента
print(ls[0], ls[-1])

# Извлечение подмассива с помощью среза
print(ls[2:4])
```

```python
import numpy as np 

# Создадим два одномерных массива
ls_1 = np.arange(1, 10)
ls_2 = np.arange(10, 19)

# Вывод поэлементного сложения
print(ls_1 + ls_2)

# Вывод поэлементного умножения
print(ls_1 * ls_2)

# Вывод средних занчений массивов
print(f"Средние величины массивов ls_1 = {np.mean(ls_1)} и ls_2 = {np.mean(ls_2)}")

# Вывод сумм массивов
print(f"Суммы массивов ls_1 = {np.sum(ls_1)} и ls_2 = {np.sum(ls_2)}")
```

```python
import numpy as np 

# Eдиничная матрица 5×5
matrix1 = np.eye(5)

print("Eдиничная матрица 5×5")
print(matrix1)
print("Прибавляем к каждому занчению +50")
print(matrix1 + 50)

# Заполняем матрицу 3x4 случайными числами
matrix2 = np.random.rand(3, 4)
print("random.randn — случайная матрица 3×4")
print(matrix2)
print("Умножаем всю матрицу на 100")
print(matrix2 * 100)

# Выполняем сложение двух матриц
print("Суммируем матрицы")
matrix3 = np.array([[1, 2, 3], 
                    [1, 2, 3]])

matrix4 = np.array([[4, 6 , 7], 
                    [1, 8, 2]])
print(matrix3 + matrix4)
```
